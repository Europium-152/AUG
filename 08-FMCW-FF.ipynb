```python
import ipfnpytools.aug_read as aug_read
import ipfnpytools.aug_plot as aug_plot
from ipfnpytools.save_figure import save_figure
```

```python
data = aug_read.fmcw_data(shot=35174)
```

```python
t_begin = 1 #1.5#3.52#2.4+0*0.38
t_end = 7 #2.5#3.84#3.35+0*1*8+0*0.38*2

fmaxs = 250
fmins = 1 #None#0
```

```python
LFS, HFS = aug_plot.fmcw(
    data=data, specgram_kw=([None] + 3 * [dict(write_fres=False, write_tres=False)]), write_f_prob=True, #write_n_e=False, 
    t_begin=t_begin, t_end=t_end, 
    f_maxs=fmaxs, f_mins=fmins,
)
```

```python
LFS_trace, HFS_trace = aug_plot.fmcw(
    data=data, plot_types='trace', y_title='(V)', write_n_e=False,
    t_begin=t_begin, t_end=t_end, 
)
```

```python
save_figure(LFS[0], LFS[2] + '_specgrams')
```
