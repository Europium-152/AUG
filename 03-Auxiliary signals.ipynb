# Auxiliary signals

Some useful signals one may use to check information about the discharges for preliminary analysis or during shot days.

```python
import matplotlib.pyplot as plt
from ipfnpytools.getsig import getsig
import numpy as np
```

## Plasma current

```python
shotnr = 35174
ipa = getsig(shotnr, 'MAG', 'Ipa')
```

```python
plt.plot(ipa.time, ipa.data)
plt.xlabel('time [s]')
plt.ylabel(ipa.unit)#Which is clearly wrong, be careful!
plt.show()
```

## Fuelling and Seeding

All fuelling information is stored in the `UVS` shotfile. Deuterium, Nitrogen, Helium, Hydrogen and Neon have their own signals. For additional seeding species such as Argon or Krypton, you must check which valve was used to inject the species.

```python
shotnr = 36448
dtot = getsig(shotnr, 'UVS', 'D_tot')
ntot = getsig(shotnr, 'UVS', 'N_tot')
artot = getsig(shotnr, 'UVS', 'CFA03A')
```

```python
plt.plot(dtot.time, dtot.data*1e-21, label='D [1e21 e/s]')
plt.plot(ntot.time, ntot.data*1e-20, label='N [1e20 e/s]')
plt.plot(artot.time, artot.data*1e-20, label='Ar [1e20 e/s]')
plt.legend()
plt.xlim(0,9)
plt.ylim(0,7)
plt.xlabel('time [s]')
plt.show()
```

## q95

There are several methods for equilibrium reconstruction at AUG, hence there are several ways q95 is calculated. In the usual current and field directions of AUG, q95 is negative.

```python
q95fpg = getsig(shotnr, 'FPG', 'q95')
q95gqi = getsig(shotnr, 'GQI', 'q95')
q95gqh = getsig(shotnr, 'GQH', 'q95')
```

```python
plt.plot(q95fpg.time, np.abs(q95fpg.data), label='q95 FPP')
plt.plot(q95fpg.time, np.abs(q95fpg.data), label='q95 EQI')
plt.plot(q95fpg.time, np.abs(q95fpg.data), label='q95 EQH')
plt.legend()
plt.xlabel('time [s]')
plt.ylim(0, 7)
plt.show()
```
