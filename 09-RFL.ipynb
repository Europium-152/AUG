```python
import ipfnpytools.aug_read as aug_read
import ipfnpytools.aug_plot as aug_plot
from ipfnpytools.plot import shadow
from ipfnpytools.save_figure import save_figure
#%matplotlib notebook
mark_hop_kw = dict(path_effects=shadow)
import matplotlib.pyplot as plt
plt.style.use('./Styles/darklab.mplstyle')
```

```python
shot = 29302
```

```python
data_Q = aug_read.hop_data(shot=shot, channel=0, return_all=True)
```

```python
data_V = aug_read.hop_data(shot=shot, channel=1, return_all=True)
```

```python
t_begin = 2
t_end = 5
mark = None#'f'#'ne'#None

fmaxs = 250
fmins = 1
```

```python
fig_Q, axes_Q, fig_name_Q = aug_plot.hop(
    data=data_Q, plot_types="specgram", y_title="Frequency (kHz)", y_labels=None, write_f_label=False,
    mark=mark, mark_hop_kw=mark_hop_kw, specgram_kw=([None] + 3 * [dict(write_fres=False, write_tres=False)]), 
    t_begin=t_begin, t_end=t_end, 
    f_mins=fmins, f_maxs=fmaxs,
)
```

```python
time_plot_Q, IQ_plot_Q = aug_plot.hop(
    data=data_Q, plot_IQ=True, t_begin=t_begin, t_end=t_end, mark=mark, crop_invalid_samples=True, plot_kw=dict(markersize=1.2))
offset_point_Q = IQ_plot_Q[1].plot(data_Q.offsets.real, data_Q.offsets.imag, 'o', color='red')
```

```python
fig_V, axes_V, fig_name_V = aug_plot.hop(
    data=data_V, plot_types="specgram", y_title="Frequency (kHz)", y_labels=None, write_f_label=False,
    mark=mark, mark_hop_kw=mark_hop_kw, specgram_kw=([None] + 3 * [dict(write_fres=False, write_tres=False)]),
    t_begin=t_begin, t_end=t_end, f_mins=fmins, f_maxs=fmaxs,
)
```

```python
time_plot_V, IQ_plot_V = aug_plot.hop(
    data=data_V, plot_IQ=True, t_begin=t_begin, t_end=t_end, mark=mark, crop_invalid_samples=True, plot_kw=dict(markersize=1.2))
offset_point_V = IQ_plot_V[1].plot(data_V.offsets.real, data_V.offsets.imag, 'o', color='red')
```

```python
save_figure(fig_Q, fig_name_Q + '_specgrams')
```
